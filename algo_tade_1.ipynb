```python
aapl = close.AAPL.copy().to_frame()
aapl 
```

```python
aapl.shift(periods=1)
```

```python
aapl["lag1"] = aapl.shift(periods=1)
```

```python
aapl #new data set so we can see the daily change in prosent
```

```python
aapl["Diff"] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl
```

```python
aapl["% Change"]=aapl.AAPL.div(aapl.lag1).sub(1).mul(100) #Finds the procent change of data from one day to the next day
aapl
```

```python
aapl["Diff2"] = aapl.AAPL.diff(periods=1)
```

```python
aapl
```

```python
aapl["% Change 2"]=aapl.AAPL.pct_change(periods=1).mul(100) #Finds the procent change of data from one day to the next day
aapl
```

```python
del aapl['Diff2'] #deletes one data frama
aapl
```

```python
del aapl['lag1'] #deletes one data frama
del aapl['Diff']
del aapl['% Change']
aapl
```

```python
aapl.rename(columns = {'% Change 2':"Change"}, inplace=True)
```

```python
ticker = ["MSFT", "AAPL", "GOOGL", "KO", "DIS", "IBM"]
stocks = yf.download(ticker, start = "2010-01-01", end="2015-01-01")

stocks
```

```python
close = stocks.loc[:,"Close"].copy() #only to the close data

norm_close = close.div(close.iloc[0]).mul(100) #normallize the data
#makes it easier to analyze the graphs 
norm_close.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show
```

```python
ret = close.pct_change().dropna()

ret.head()
```

```python
ret.describe().T
```

```python
summary = ret.describe().T.loc[:,["mean", "std"]]
summary
```

```python
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std", y="mean", figsize=(12,8), s=50, fontsize=15)
for i in summary.index:
  plt.annotate(i, xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002), size=15) #Just used to label graph
plt.xlabel("Annual risk(std)", fontsize=15)
plt.ylabel("Annual return", fontsize=15)
plt.title("Risk/return", fontsize=25)
plt.show()
```

Correlation and coveriance

```python
ret.cov()

```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(ret.corr(), cmap="Reds", annot=True, annot_kws={"size":15}, vmax=0.6) #Show the correlation inbetween the industries etc
plt.show
```
